```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate random data following a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(12,5))
sns.histplot(data, bins=30, kde=True, color='skyblue')
plt.title("Normal Distribution", fontsize=14, weight='bold')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Binomial distribution example
n = 10   # number of trials
p = 0.5  # probability of success

data = np.random.binomial(n=n, p=p, size=1000)

plt.figure(figsize=(8,5))
sns.histplot(data, bins=range(0, n+2), color='lightgreen', edgecolor='black')
plt.title("Binomial Distribution", fontsize=14, weight='bold')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# Uniform distribution
data = np.random.uniform(low=10, high=50, size=1000)

plt.figure(figsize=(12,5))
sns.histplot(data, bins=20, color='orange', edgecolor='black')
plt.title("Uniform Distribution", fontsize=14, weight='bold')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
