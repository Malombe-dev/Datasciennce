```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data/sleep.csv")


```

```python
plt.figure(figsize=(10,5))

# Use the unique sorted values as bins
bins = sorted(df["Sleep Duration"].unique())

plt.hist(df["Sleep Duration"], bins=bins, color="skyblue", edgecolor="black", alpha=0.7)

plt.title("Sleep Duration Distribution (per 0.1 hr)", fontsize=14, weight="bold")
plt.xlabel("Hours of Sleep")
plt.ylabel("Number of People")
plt.xticks(bins, rotation=45)   # show all bins on x-axis
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Sleep Duration", data=df, palette="Set2")

plt.title("Sleep Duration by Gender", fontsize=14, weight="bold")
plt.xlabel("Gender")
plt.ylabel("Hours of Sleep")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Sleep Duration", data=df, hue="Gender", alpha=0.7)

plt.title("Age vs Sleep Duration", fontsize=14, weight="bold")
plt.xlabel("Age")
plt.ylabel("Hours of Sleep")
plt.show()
```
