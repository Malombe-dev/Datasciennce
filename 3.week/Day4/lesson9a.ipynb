```python
import pandas as pd
df = pd.read_csv("data/sleep.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)   # correlation of numeric columns
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, linewidths=0.5)

plt.title(" Correlation Heatmap of Sleep Health Data", fontsize=14, weight="bold")
plt.show()
```

```python
sns.pairplot(df, vars=["Sleep Duration", "Quality of Sleep", "Stress Level"], hue="Gender")
plt.suptitle(" Pairwise Relationships in Sleep Data", y=1.02, fontsize=14, weight="bold")
plt.show()
```

```python
sns.catplot(x="Occupation", y="Sleep Duration", kind="box", hue="Gender", data=df, height=6, aspect=1.5)
plt.title(" Occupation vs Sleep Duration by Gender", fontsize=14, weight="bold")
plt.xticks(rotation=45)
plt.show()

```
